### Проверка расчётов эффективности обогащения

Проверьте, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака rougher.output.recovery. Найдите MAE между вашими расчётами и значением признака. Опишите выводы.

Эффективность обогащения: Recovery = C*(F-T)/F*(C-T) * 100%

где: 
C — доля золота в концентрате после флотации/очистки;
F — доля золота в сырье/концентрате до флотации/очистки;
T — доля золота в отвальных хвостах после флотации/очистки.


```python
def recovery(C, F, T):
    numerator = C * (F - T)
    denominator = F * (C - T)
    rec = numerator / denominator * 100
    
    rec[rec<0] = np.nan
    rec[rec>100] = np.nan

    return rec
```

```python
c, f, t = train['rougher.output.concentrate_au'], train['rougher.input.feed_au'], train['rougher.output.tail_au']
rough_recov_calc = recovery(c, f, t)

rough_recov_train = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.output.recovery']

right_rough_recov_train = np.isclose(train['rougher.output.recovery'], rough_recov_calc).sum()

# совпадения и не совпадения значений
right_rough_recov_train,train.shape[0]-right_rough_recov_train
```

Между посчитанным значением recovery для дф train и изначальными значениями параметра 2489 отличия

**Посчитаем MAE между нашими значениями rec и изначальными**

```python
mean_absolute_error(train['rougher.output.recovery'].interpolate(method='time'),
                    rough_recov_calc.interpolate(method='time'))
```

Средняя асолютная ошибка между данными очень мала, поэтому можем считать, что recovery посчитан верно

```python
display(rough_recov_calc.isna().sum())
display(train['rougher.output.recovery'].isna().sum())
```

Пропуски равны в данных, значит, скорее всего изначально данные не были заполнены при создании дф.

Создадим функцию, которая покажет какие параметры отсутствуют и в каком количество

```python
def show_na(data):
    na_info = data.isna().sum()
    results = (pd.concat([na_info / data.shape[0], na_info], axis=1, keys=['percent', 'count'])
        .sort_values('percent', ascending=False))
    results['percent'] = results['percent'].apply(lambda x: f'{x*100:.2f}%')
    return results
```

### Расчет метрики sMAPE

Напишем функцию для вычисления итоговой sMAPE.

Итоговое sMAPE: 


**sMAPE rougher**

```python
def sMAPE(y_true, y_pred):
    
    numerator = abs(y_true-y_pred)
    denominator =(abs(y_true) + abs(y_pred))/2    
    N = len(y_true)
    smae_rough_rec = 1 / N * (numerator/denominator).sum() * 100

    print("Показатель sMAPE для 'Эффективности обогащения золота на этапе флотации'")
    print("между рассчитанным значением и тренировочной выборкой: {0:.2E}".format(smae_rough_rec))

smape_r = sMAPE(rough_recov_train, rough_recov_calc)

```

```python
def smape(true, predict):
    return 1 / len(true) * sum( abs(true - predict) / ( (abs(true) + abs(predict)) * 0.5) ) * 100
```

```python
def final_smape(smape_r, smape_f):
    return 0.25 * smape_r + 0.75 * smape_f
```
